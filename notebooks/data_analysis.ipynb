## Netflix Series Data Analysis

### Import libraries

```python
import sys
import os
sys.path.append('../scripts')

from queries import fetch_data, load_query
import matplotlib.pyplot as plt
import seaborn as sns

path = os.path.join(os.path.dirname(os.getcwd()), 'scripts', 'queries.sql')
```

### Netflix series titles per country

```python
query = load_query("QUERY_TITLES_PER_COUNTRY", path)
df = fetch_data(query)
```

```python
plt.figure(figsize=(12,6))
sns.barplot(data=df.head(10), x='country', y='total_titles', palette='flare', hue='country', legend=False)
plt.title("Number of titles per country (Top 10)", fontsize=16)
plt.xlabel('Country', fontsize=12)
plt.ylabel('Number of titles', fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
