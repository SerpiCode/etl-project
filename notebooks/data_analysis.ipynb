## Netflix Series Data Analysis

### Import libraries

```python
import sys
import os
sys.path.append('../scripts')

from queries import fetch_data, load_query
import matplotlib.pyplot as plt
import seaborn as sns

path = os.path.join(os.path.dirname(os.getcwd()), 'scripts', 'queries.sql')
```

### Netflix series titles per country

```python
query = load_query("QUERY_TITLES_PER_COUNTRY", path)
df = fetch_data(query)
```

```python
plt.figure(figsize=(12,6))
sns.barplot(data=df.head(10), x='country', y='total_titles', palette='flare', hue='country', legend=False)
plt.title("NUMBER OF TITLES PER COUNTRY (TOP 10)", fontsize=16, weight='bold')
plt.xlabel('Country', fontsize=12)
plt.ylabel('Number of titles', fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

### Number of titles added per year

```python
query = load_query("QUERY_TITLES_BY_YEAR", path)
df = fetch_data(query)
```

```python
plt.figure(figsize=(12,6))
sns.lineplot(data=df, x='year', y='total_titles', markers='o')
plt.title('NUMBER OF TITLES ADDED PER YEAR', fontsize=16, weight='bold')
plt.xlabel('Year', fontsize=12)
plt.ylabel('Number of titles', fontsize=12)
plt.grid(True)
plt.tight_layout()
plt.show()
```

### Number of titles per rating

```python
query = load_query("QUERY_TITLES_BY_RATING", path)
df = fetch_data(query)
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(data=df, x='rating', y='total', palette='coolwarm', hue='rating', legend=False)
plt.title("TITLE'S RATING DISTRIBUTION", fontsize=16, weight='bold')
plt.xlabel("Rating", fontsize=12)
plt.ylabel("Number of titles", fontsize=12)
plt.tight_layout()
plt.show()
```
